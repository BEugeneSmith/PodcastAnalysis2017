```python
from Analysis.Helper import *
gen = datetime.now()
start = datetime(2019,1,1)
end = datetime(2019,7,1)
aai = AudioAnalysisInput(start=start,end=end)
```

```python
data = aai.data[(aai.data['datePlayed']>=start)]
data = data[(~data['batch_skip_ind'])|(~data['single_skip_ind'])]
filt_data = data[['weekPlayed','album','duration','category','datePlayed']]
# filt_data.to_clipboard(index=False)
```

```python
stacked_chart_data = (filt_data.pivot_table(index="weekPlayed",columns="category",values="duration").fillna(0)/60).apply(np.ceil)
stacked_chart_data.index = list(map(lambda x: int(x[-2:]),stacked_chart_data.index))
stacked_chart_data.to_csv(gen.strftime('stacked_chart_data_%Y%m%d.csv'))
```
