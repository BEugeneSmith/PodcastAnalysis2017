```python
import datetime,glob,os
import pandas as pd
from pandas.tseries.holiday import USFederalHolidayCalendar
```

```python
daystocheck = 365
audio_type="Podcast"
today = datetime.datetime.today()
weekago = today - datetime.timedelta(daystocheck)
_holidays = USFederalHolidayCalendar()

for i in range(daystocheck):
    print(weekago)
    p_weekago = weekago.strftime("data/{}ListeningPull%Y%m%d.txt".format(audio_type))
    missing_ind = not os.path.exists(p_weekago)
    if missing_ind:
        print("reportdate:")
        print(weekago)
        date_cols = ['datePlayed','dateAdded','dateReleased']
        data = pd.read_table('alldata20190629.txt',sep='|',parse_dates=date_cols,error_bad_lines=False)
#         data = pd.read_table('scripts/all_data_dump20190520.txt',sep='|',parse_dates=date_cols,error_bad_lines=False)
        
        data.loc[:,'reportRunDate'] = weekago
        
        data.loc[:,'expectedDatePlayed'] = data['reportRunDate'] - datetime.timedelta(days=1)
        data.loc[:,'played_yesterday_ind'] = pd.to_datetime(data['datePlayed']).dt.strftime('%Y%m%d') == pd.to_datetime(data['expectedDatePlayed']).dt.strftime('%Y%m%d')
        print(data.shape)
        data = data[data['played_yesterday_ind']==True]
        print(data.shape)
        data.drop(['played_yesterday_ind','expectedDatePlayed'],inplace=True,axis=1)
        
        # create indicators for holiday and weekend play
        data.loc[:,'played_weekend_ind'] = data['datePlayed'].apply(lambda x: x.weekday() in [5,6])
        data.loc[:,'played_holiday_ind'] = data['datePlayed'].apply(lambda x: x.weekday() in _holidays.holidays())

        # pre-calclulate critical time diffs in seconds
        data['release_to_play_seconds'] = (data['datePlayed'] - pd.to_datetime(data['dateReleased'])).dt.total_seconds()
        data['release_to_add_seconds'] = (data['dateAdded'] - pd.to_datetime(data['dateReleased'])).dt.total_seconds()
        data['add_to_play_seconds'] = (data['datePlayed'] - data['dateAdded']).dt.total_seconds()
 
        project_dir = '/Users/BEugeneSmith/Desktop/projects/PodcastAnalysis'
#         print(p_weekago)
#         out_file_name = p_weekago.strftime('data/{}ListeningPull%Y%m%d.txt'.format(audio_type))
        data.to_csv(os.path.join(project_dir,p_weekago),sep='|',index=False)

    weekago += datetime.timedelta(1)
```
