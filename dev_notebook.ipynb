```python
import pandas as pd
```

```python
date_cols = ['datePlayed','dateAdded','dateReleased']
data = pd.read_table('data/PodcastListeningPull20180108.txt',sep='|',parse_dates=date_cols)
```

```python
data['release_to_play_seconds'] = (data['datePlayed'] - data['dateReleased']).dt.total_seconds()
data['release_to_add_seconds'] = (data['dateAdded'] - data['dateReleased']).dt.total_seconds()
data['add_to_play_seconds'] = (data['datePlayed'] - data['dateAdded']).dt.total_seconds()
```

```python
data.sort_values('add_to_play_seconds').tail()
```
