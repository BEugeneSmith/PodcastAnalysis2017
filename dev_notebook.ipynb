```python
import pandas as pd
import json
```

```python
moment_file = 'moment.json'
```

```python
def load_clean_json(moment_json):
    buffer = open(moment_json)

    all_lines = buffer.readlines()
    all_lines = ''.join(all_lines)
    all_lines = all_lines.replace('\n','')
    return json.loads(all_lines)
```

```python
def parse_json_to_dataframe(json_data):
    json_dict = json_data['days']
    pickup_fields = ['date', 'locationAccuracyInMeters', 'longitude', 'latitude']
    day_labels = ['date','pickups']
    records = {
        'real_datetime':[],
        'pickup_datetime':[],
        'location_accuracy':[],
        'longitude':[],
        'latitude':[]
    }

    for data_ix in range(len(json_dict)):
        real_dt = ''
        for label in day_labels:
            label_data = json_dict[data_ix][label]
            if label =='date':
                real_dt = label_data
            else:
                pickups = label_data
                for pickup in pickups:
                    records['pickup_datetime'].append(pickup['date'])
                    records['location_accuracy'].append(pickup['locationAccuracyInMeters'])
                    records['longitude'].append(pickup['longitude'])
                    records['latitude'].append(pickup['latitude'])
                    records['real_datetime'].append(real_dt)
    return pd.DataFrame(records)
```

```python
json_data = load_clean_json(moment_file)
parse_json_to_dataframe(json_data)
```
